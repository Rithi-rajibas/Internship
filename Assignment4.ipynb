1. Scrape the details of most viewed videos on YouTube from Wikipedia. Url 
= https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos You need to find following details: A)
Rank 
B) Name 
C) Artist 
D) Upload date 
E) Views

```python
driver=webdriver.Chrome()
driver.maximize_window()
driver.get("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos")
```

```python


VideoName=[]
Uploader=[]
Upload_Date=[]
Views=[]

#Scraping Video Name
videoName=driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"][1]/tbody/tr/td[1]')

try:
    for i in videoName:
         VideoName.append(i.text)
except NoSuchElementException:
    VideoName.append('-')
        

#Scraping Uploader
uploader=driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"][1]/tbody/tr/td[2]')
try:
    for i in uploader:
        Uploader.append(i.text)
except NoSuchElementException:
    Uploader.append('-')
    


#Scraping Video views     
views=driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"][1]/tbody/tr/td[3]')
try: 
    for i in views:
        Views.append(i.text)
except NoSuchElementException:
    Views.append('-')
    
    

#Scraping upload Date
date=driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"][1]/tbody/tr/td[4]')
try:
    for i in date:
        Upload_Date.append(i.text)
except NoSuchElementException:
    Upload_Date.append('-')
    
```

```python
print(len(VideoName),len(Uploader),len(Views),len(Upload_Date))
```

```python
#Creating DataFrame for Wikidata
import pandas as pd
WikiData=pd.DataFrame({"Video Name":VideoName,"Date":Upload_Date,
                       "Uploader":Uploader,"Views":Views})
WikiData
```

8. Details of Datasets from UCI machine learning repositories. 
Url = https://archive.ics.uci.edu/ You 
have to find the following details: 
A) Dataset name 
B) Data type 
C) Task 
D) Attribute type 
E) No of instances 
F) No of attribute G) Year 
 Note: - from the home page you have to go to the Show All Dataset page through code

```python
driver=webdriver.Chrome()
driver.maximize_window()
driver.get("https://archive.ics.uci.edu/")
```

```python
#click on view dataset
view_dataset_link=driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[1]/div/div/div/a[1]')
view_dataset_link.click()
```

```python
 #Click on Expand all link
Expand_All_lnk=driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[2]/div[1]/div/label[2]/div[2]/span[2]')
Expand_All_lnk.click()
```

```python
Dataset_Name=[]
Data_type=[]
Task=[]
Attribute_type=[]
Instances=[]
Attribute_No=[]
Year=[]

  
for i in range(60):
 #Click on Collapse all link 
        collapse_lnk=driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[2]/div[1]/div/label[2]/div[2]/span[1]')
        collapse_lnk.click()
 #Click on Expand all link
        Expand_All_lnk=driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[2]/div[1]/div/label[2]/div[2]/span[2]')
        Expand_All_lnk.click()
        
    #Scraping Dataset Name
        DS_Name=driver.find_elements(By.XPATH,'//a[@class="link-hover link text-xl font-semibold"]')
        try:
            for i in DS_Name:
                Dataset_Name.append(i.text)
        except NoSuchElementException:
            Dataset_Name.append('-')
            
    #Scraping DataType Name
        datatype=driver.find_elements(By.XPATH,'//div[@class="my-2 hidden gap-4 md:grid grid-cols-12"]/div[2]')
        try:
            for i in datatype:
                Data_type.append(i.text)
        except NoSuchElementException:
            Data_type.append('-')
    #Scraping Task Name
        task=driver.find_elements(By.XPATH,'//p[@class="truncate"]')
        try:
            for i in task:
                Task.append(i.text)
        except NoSuchElementException:
            Task.append('-')
            
     #Scraping Attribute type
        atype=driver.find_elements(By.XPATH,'//tbody[@class="border"]/tr/td[2]')
        try:
            for i in atype:
                 Attribute_type.append(i.text)
        except NoSuchElementException:
                Attribute_type.append('-')
                
    #Scraping No Of Instances
        instances=driver.find_elements(By.XPATH,'//div[@class="my-2 hidden gap-4 md:grid grid-cols-12"]/div[3]')
        try:
            for i in instances:
                Instances.append(i.text)
        except NoSuchElementException:
            Instances.append('-')
            
    #Scraping No Of Attributes
        attributes= driver.find_elements(By.XPATH,'//div[@class="my-2 hidden gap-4 md:grid grid-cols-12"]/div[4]')
        try:
             for i in attributes:
                    Attribute_No.append(i.text)
        except NoSuchElementException:
             Attribute_No.append('-')
    
    #Scraping Year
        year=driver.find_elements(By.XPATH,'//tbody[@class="border"]/tr/td[3]')
        try:
            for i in year:
                    Year.append(i.text)
        except NoSuchElementException:
            Year.append('-')
try:           
    next_button=wait.until(EC.element_to_be_clickable((By.XPATH,"/html/body/div/div[1]/div[1]/main/div/div[2]/div[3]/div/button[2]")))
    next_button.click()
except NoSuchWindowException as e:
    print("Exception Raised Message",e)
    
```

```python
print(len(Dataset_Name),len(Data_type),len(Task),len(Attribute_type),len(Instances),len(Attribute_No),len(Year))
```

```python
import pandas as pd
Dataset=pd.DataFrame({"Dataset Name":Dataset_Name,"Data Type":Data_type,"Task":Task,"Attribute Type":Attribute_type,
                      "No of Instances":Instances,"No Of Attribute":Attribute_No,"Year":Year})
Dataset
```
