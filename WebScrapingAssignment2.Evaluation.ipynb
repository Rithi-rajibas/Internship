Q1: In this question you have to scrape data using the filters available on the webpage You have to use the location and 
salary filter. 
You have to scrape data for ‚ÄúData Scientist‚Äù designation for first 10 job results. 
You have to scrape the job-title, job-location, company name, experience required. 
The location filter to be used is ‚ÄúDelhi/NCR‚Äù. The salary filter to be used is ‚Äú3-6‚Äù lakhs

```python
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
import time
```

```python
driver=webdriver.Chrome()
driver.get("https://www.naukri.com/")
driver.maximize_window()
```

```python
#Entering Designation,Job position

Designation=driver.find_element(By.CLASS_NAME,"suggestor-input ")
Designation.send_keys("Data Scientist")

search=driver.find_element(By.CLASS_NAME,"qsbSubmit")
search.click()
```

```python
#Filter for Location and Salary
Salary=driver.find_element(By.XPATH,'/html/body/div/div/main/div[1]/div[1]/div/div/div[2]/div[2]/div[2]/div[2]/label/p/span[1]')
Salary.click()
```

```python
locationfilter=driver.find_element(By.XPATH,'/html/body/div/div/main/div[1]/div[1]/div/div/div[2]/div[6]/div[2]/div[3]/label/p/span[1]')
locationfilter.click()
```

```python
#Scraping Jobtitle,Location,Company_Name,Experience_Required

Jobtitle=[]
Job_Location=[]
company_name=[]
Experience_required=[]

titletags=driver.find_elements(By.XPATH,'//div[@class="cust-job-tuple layout-wrapper lay-2 sjw__tuple "]/div/a')
for i in titletags[:10]:
    title=i.text
    Jobtitle.append(title)
    
Locationtags=driver.find_elements(By.XPATH,'//span[@class="locWdth"]')
for i in Locationtags[:10]:
    Location=i.text
    Job_Location.append(Location)
    
Companytags=driver.find_elements(By.XPATH,'//div[@class=" row2"]/span/a[1]')
for i in Companytags[:10]:
    Company=i.text
    company_name.append(Company)
    
ExpTags=driver.find_elements(By.XPATH,'//span[@class="expwdth"]')
for i in ExpTags[:10]:
    Exp=i.text
    Experience_required.append(Exp)
```

```python
print(len(Jobtitle),len(Job_Location),len(company_name),len(Experience_required))
```

```python
#Naukri Data
import pandas as pd
NaukriData=pd.DataFrame({"Title":Jobtitle,'Location':Job_Location,'Company Name':company_name,'Experience':Experience_required})
NaukriData
```

Q6: Write a python program to scrape data for Top 1000 Quotes of All Time.
The above task will be done in following steps:
1. First get the webpagehttps://www.azquotes.com/
2. Click on TopQuote
3. Than scrap a)Quote b) Author c) Type Of Quotes

```python
driver=webdriver.Chrome()
driver.maximize_window()
driver.get("https://www.azquotes.com/")
```

```python
#Click on TopQuote
TopQuotelink=driver.find_element(By.XPATH,"/html/body/div[1]/div[1]/div[1]/div/div[3]/ul/li[5]/a")
TopQuotelink.click()
```

```python
#Scraping Data for Quotes,Author and Type of Quotes
page=[]
Quotes=[]
Author=[]
TypeOfQuotes=[]

for page in range(0,10):
    quote=driver.find_elements(By.XPATH,'//div[@class="wrap-block"]/p')
    for i in quote:
        Quotes.append(i.text)
    
    auth=driver.find_elements(By.XPATH,'//div[@class="author"]/a')
    for i in auth:
        Author.append(i.text)
        
    type=driver.find_elements(By.XPATH,'//div[@class="tags"]')
    for i in type:
        TypeOfQuotes.append(i.text)
        
        
next_button=driver.find_element(By.XPATH,"/html/body/div[1]/div[2]/div/div/div/div[1]/div/div[4]/li[12]")
next_button.click()
time.sleep(3)

    
```

```python
print(len(Quotes),len(Author),len(TypeOfQuotes))
```

```python
#DataFrame for top 1000 Quotes
import pandas as pd
TopQuotes=pd.DataFrame({"Quotes":Quotes,"Author":Author,"Type of Quote":TypeOfQuotes})
TopQuotes
```

Q7: Write a python program to display list of respected former Prime Ministers of India (i.e. Name,
Born-Dead, Term of office, Remarks) from https://www.jagranjosh.com/general-knowledge/list-ofall-prime-ministers-of-india-1473165149-1
scrap the mentioned data and make the DataFrame

```python
driver=webdriver.Chrome()
driver.maximize_window()
driver.get("https://www.jagranjosh.com/general-knowledge/list-ofall-prime-ministers-of-india-1473165149-1")

```

```python
#Scraping Name,Born_dead ,Terms OF Office and Remarks of PrimeMinister list
Names=[]
Born_Dead=[]
TermOfOffice=[]
Remarks=[]


Name=driver.find_elements(By.XPATH,'//div[@class="TableData"][1]/table/tbody/tr/td[2]')
for i in Name:
    Name=i.text
    Names.append(Name)
    
BornDead=driver.find_elements(By.XPATH,'//div[@class="TableData"][1]/table/tbody/tr/td[3]')
for i in BornDead:
    Born=i.text
    Born_Dead.append(Born)
    
Term=driver.find_elements(By.XPATH,'//div[@class="TableData"][1]/table/tbody/tr/td[4]')
for i in Term:
    Terms=i.text
    TermOfOffice.append(Terms)
    
Remark=driver.find_elements(By.XPATH,'//div[@class="TableData"][1]/table/tbody/tr/td[5]')
for i in Remark:
    remark=i.text
    Remarks.append(remark)

    
```

```python
print(len(Names),len(Born_Dead),len(TermOfOffice),len(Remarks))
```

```python
#DataFrame For PrimeMinister
import pandas as pd
PrimeMinistersOfIndia=pd.DataFrame({"Names":Names[:19],"Born_Dead":Born_Dead[:19],"Terms-OF-Office":TermOfOffice,"Remarks":Remarks})
PrimeMinistersOfIndia
```
